```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

df=pd.read_csv("creditcard.csv")
df
```

```python
selected_columns=["V18","V19","V20","V21","V22","V23","V24","V25","V26","V27","V28","Amount","Class"]
```

```python
selected_data=df[selected_columns]
```

```python
print(selected_data)
```

```python
selected_data=df[selected_columns].astype(bool)
mean_value = df[selected_data].mean()
df[selected_data] = df[selected_data].fillna(mean_value)
```

```python
df.isnull().sum()
```

```python
features = df[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
target = df['Class']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)

score = model.score(X_test, y_test)
print('Accuracy:', score)
```

```python
predictions = model.predict(X_test)

print('Predictions:', predictions)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
